```python
# imports
import pandas as pd

data = pd.read_csv('C:/Users/yacoo/DataMiningCourse/Assignments/Assignment-Three/aaup.csv', header=None)
data.columns = ['Univ_id', 'Univ_name', 'State', 'Type', 'fp_sal', 'ac_sal', 'at_sal', 'to_sal', 'fp_com', 'ac_com', 'at_com', 'to_com', 'fp_#', 'ac_#', 'at_#', 'in_#', 'to_#']

# contains the row
print('Number of observations = %d' % (data.shape[0]))
# contains the column
print('Number of attributes = %d' % (data.shape[1]))

# show the first 100 values across all attributes
data.head(100)
```

```python
import numpy as np

# recode missing values of '*' into 'NaN'
data = data.replace('*', np.NaN)

# contains the row
print('Number of observations = %d' % (data.shape[0]))
# contains the column
print('Number of attributes = %d' % (data.shape[1]))

print('Number of missing values across the attributes:')
for col in data.columns:
    print('\t%s: %d' % (col,data[col].isna().sum()))
```

```python
# list
columns_with_missing = ['fp_sal', 'ac_sal', 'at_sal', 'fp_com', 'ac_com', 'at_com']

print('Before dropping missing values:')
for col in columns_with_missing:
    print(data[col][35:40])
    
print('\nNumber of rows in original data = %d' % (data.shape[0]))
data.dropna(subset=columns_with_missing, inplace=True)

print('\nAfter dropping missing values:')
for col in columns_with_missing:
    print(data[col][35:40])

print('\nNumber of rows after dropping missing values = %d' % (data.shape[0]))
```

```python
"""
for some odd reason, numeric attributes were not being 'detected' as they should have 
and therefore were converted to that
"""
data['fp_sal'] = pd.to_numeric(data['fp_sal'])

# histogram scaled out that shows all the data
print(data['fp_sal'].value_counts(sort=False))
data['fp_sal'].hist(bins=10)
```

```python
# discretize the attributes using equal width method
bins = pd.cut(data['fp_sal'], 5)
print(bins.value_counts(sort=False))
# showing what it would mean to discretize the bins into categorical attributes
data['fp_sal'].hist(bins=5)
```

```python
# discretize the attributes using equal frequency method
bins = pd.qcut(data['fp_sal'], 3)
print(bins.value_counts(sort=False))
# showing what it would mean to discretize the bins with nearly same number of instances
data['fp_sal'].hist(bins=3)
```

```python
data['fp_#'] = pd.to_numeric(data['fp_#'])

# histogram scaled out that shows all the data
print(data['fp_#'].value_counts(sort=False))
data['fp_#'].hist(bins=10)
```

```python
bins = pd.cut(data['fp_#'], 5)
print(bins.value_counts(sort=False))
data['fp_#'].hist(bins=5)
```

```python
bins = pd.qcut(data['fp_#'], 3)
print(bins.value_counts(sort=False))
data['fp_#'].hist(bins=3)
```
