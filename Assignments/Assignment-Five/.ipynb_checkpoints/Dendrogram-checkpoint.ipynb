```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Define your similarity values and cluster names
similarity_matrix = np.array([[1,.1,.41,.55,.35],
                              [.1,1,.64,.47,.98],
                              [.41,.64,1,.44,.85],
                              [.55,.47,.44,1,.76],
                              [.35,.98,.85,.76,1]])

cluster_names = ['1', '2', '3', '4', '5']

# Create a linkage matrix from the similarity matrix
linkage_matrix = linkage(similarity_matrix, method='complete')

# Plot the dendrogram
dendrogram(linkage_matrix, labels=cluster_names)
plt.ylabel('Similarity')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Define your similarity values and cluster names
similarity_matrix = np.array([[1,.1,.41,.55,.35],
                              [.1,1,.64,.47,.98],
                              [.41,.64,1,.44,.85],
                              [.55,.47,.44,1,.76],
                              [.35,.98,.85,.76,1]])

cluster_names = ['1', '2', '3', '4', '5']

# Create a linkage matrix from the similarity matrix
linkage_matrix = linkage(similarity_matrix, method='single')

# Plot the dendrogram
dendrogram(linkage_matrix, labels=cluster_names)
plt.ylabel('Similarity')
plt.show()
```
