```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
iris = datasets.load_iris()
X = iris.data

# DataFrame
iris_df = pd.DataFrame(X, columns=iris.feature_names)

# K = 2 to K = 5
K_VALUES = [2, 3, 4, 5]

SSE = []
cluster_assignments = []

# Perform K-means clustering for each K and visualize the results
for k in K_VALUES:
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    
    # Get cluster assignments for each tuple
    labels = k_means.labels_
    cluster_assignments.append(labels)
    
    sse = k_means.inertia_
    SSE.append(sse)
    
    # Visualize the clusters
    plt.figure(figsize=(8, 6))
    colors = plt.cm.nipy_spectral(labels.astype(float) / k)
    plt.scatter(X[:, 0], X[:, 1], c=colors, s=50)
    plt.title(f'K-means Clustering with K={k}')
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    
    # Plot
    centroids = k_means.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, label='Centroids')
    
    plt.legend()
    plt.show()

# Plot SSE vs. Number of Clusters
plt.plot(K_VALUES, SSE)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('SSE')
plt.show()
```
